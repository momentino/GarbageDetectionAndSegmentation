### Crop images to have an aspect ratio of 4:3 (the same as our dataset)

```python
negative_img_list_path = '/home/filippo/Downloads/gtos/gtos/color_imgs'
subdirs = [x[0] for x in os.walk(negative_img_list_path)]                                                                            
for subdir in subdirs:                                                                                            
    files = os.walk(subdir).__next__()[2]                                                                             
    if (len(files) > 0):                                                                                          
        for file in files:    
            try:
                second_img = False
                print(file)
                img_path = os.path.join(subdir,file)
                im = Image.open(img_path)
                width, height = im.size   # Get dimensions

                if(width <= height):
                    new_width = width
                    new_height = 0.75*width
                    second_img = True
                    top = height - new_height
                else:
                    new_width = height/0.75
                    new_height = height
                    top = 0
                print("Width ",width)
                print("Height ",height)
                left = 0
                
                right = new_width
                bottom = height

                # Crop 
                im1 = im.crop((left, top, right, bottom))
                if(".jpg" not in img_path):
                    im1.save(img_path+".jpg")
                else:
                    im1.save(img_path)

                if(second_img == True):
                    top = 0
                    bottom = new_height
                    im2 = im.crop((left, top, right, bottom))
                    im2.save(img_path+"_1.jpg")
            except:
                print("errore")
            
```

### Delete all files with no extention

```python
for filename in os.listdir(negative_img_list_path):
    file_path = os.path.join(negative_img_list_path, filename)
    try:
        if (os.path.isfile(file_path) or os.path.islink(file_path)) and ".jpg" not in file_path:
            os.unlink(file_path)
    except Exception as e:
        print('Failed to delete %s. Reason: %s' % (file_path, e))
```

### Copy images from GTOS dataset into a single folder

```python
folder = r"/home/filippo/Downloads/gtos/gtos/color_imgs"

subfolders = [f.path for f in os.scandir(folder) if f.is_dir()]

for sub in subfolders:
    for f in os.listdir(sub):
        src = os.path.join(sub, f)
        dst = os.path.join("/home/filippo/Downloads/archive/garbage classification/negative", f)
        shutil.move(src, dst)
```
