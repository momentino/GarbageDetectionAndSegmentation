```python
import requests
from PIL import Image
import os
```

### Get negative images from the TACO dataset

```python
negative_img_list_path = "/home/momentino/TACO/detector/taco_config/background_imgs"
iterator = 0
with open(negative_img_list_path) as topo_file:
    for image_url in topo_file:  
        img_data = requests.get(image_url).content
        with open("/home/momentino/Scaricati/archive/garbage classification/Garbage classification/negative/taco_"+str(iterator), 'wb') as handler:
            handler.write(img_data)
        iterator = iterator+1
```

### Normalize names of images included in the folder (TACO + IndoorOutdoorDataset)

```python
negative_img_path = "/home/momentino/Scaricati/archive/garbage classification/Garbage classification/negative"
for i, filename in enumerate(os.listdir(negative_img_path)):
    print(filename,i)
    os.rename(os.path.join(negative_img_path,filename), os.path.join(negative_img_path,"negative_"+str(i)))
```

### Crop images to have an aspect ratio of 4:3 (the same as the garbage dataset)

```python
negative_img_list_path = '/home/momentino/Scaricati/archive/garbage classification/Garbage classification/negative'
subdirs = [x[0] for x in os.walk(negative_img_list_path)]                                                                            
for subdir in subdirs:                                                                                            
    files = os.walk(subdir).__next__()[2]                                                                             
    if (len(files) > 0):                                                                                          
        for file in files:    
            try:
                second_img = False
                print(file)
                img_path = os.path.join(subdir,file)
                im = Image.open(img_path)
                height, width = im.size   # Get dimensions

                if(width < height):
                    new_width = width
                    new_height = 0.75*width
                    second_img = True
                else:
                    new_width = height/0.75
                    new_height = height
                print("Width ",width)
                print("Height ",height)
                left = 0
                top = height - new_height
                right = new_width
                bottom = height

                # Crop 
                im1 = im.crop((left, top, right, bottom))
                if(".jpg" not in img_path):
                    im1.save(img_path+".jpg")
                else:
                    im1.save(img_path)

                if(second_img == True):
                    top = 0
                    bottom = new_height
                    im2 = im.crop((left, top, right, bottom))
                    im2.save(img_path+"_1.jpg")
            except:
                print("errore")
            
```
